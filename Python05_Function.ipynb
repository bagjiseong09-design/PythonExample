```python
def my_gen():
    try:
        yield 1
        yield 2
        yield 3
    finally:
        print("Generator closed")


gen = my_gen()
print(next(gen))
gen.close()
```

```python
def echo_generator():
    while True:
        received = yield
        print("Received:", received)

gen = echo_generator()
next(gen)  # Prime the generator

gen.send("Hello")
gen.send("World")
```

```python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

# Get first 100 Fibonacci numbers
gen = fibonacci()

for _ in range(100):
    print(next(gen))
```

```python
def large_sequence(n):
    for i in range(n):
        yield i

# This doesn't create a million numbers in memory
gen = large_sequence(1000000)

print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

for num in count_up_to(5):
    print(num)
```

```python
import functools

def changecase(func):
    @functools.wraps(func)
    def myinner():
        return func().upper()
    return myinner

@changecase
def myfunction():
    return "Have a great day!"

print(myfunction.__name__)
print(myfunction())
```
